```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
train=pd.read_csv("fraudTrain.csv")
test=pd.read_csv("fraudTest.csv")
```

```python
def clean_data(clean):
    clean.drop(["Unnamed: 0",'cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],axis=1, inplace=True)
    clean.dropna()
    return clean
```

```python
clean_data(train)
```

```python
clean_data(test)
```

```python
#mapping the values of gender to binary values
train['gender']=train['gender'].map({'M':0,'F':1})
```

```python
#encoding function for all the values of the objects(merchant,category,job and gender) since they are strings and alpha characters
enc=LabelEncoder()
def encode(d):
    d['merchant']=enc.fit_transform(d['merchant'])
    d['category']=enc.fit_transform(d['category'])
    d['gender']=enc.fit_transform(d['gender'])
    d['job']=enc.fit_transform(d['job'])
    return d
```

```python
#encoding values of train data
encode(train)
```

```python
#encoding values of test data
encode(test)
```

```python
train.corr()
```

```python
# Compute the correlation matrix
corr = train.corr()

# Plot the correlogram
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='viridis', fmt=".2f", cbar_kws={"shrink": .8}, linewidths=0.5, square=True)
plt.title('Correlogram')
plt.show()
```

```python
#assigning values to x and y
x = train.drop(columns=["is_fraud"], inplace = False)
y = train["is_fraud"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#applying logistic regression
log_reg = LogisticRegression(max_iter=1000)
```

```python
log_reg.fit(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
from sklearn.metrics import classification_report
# Evaluating and printing the performance metrics
accuracy1 = accuracy_score(y_test, y_pred)
print(f"Accuracy by logistic regression = {(accuracy1 * 100):.2f}%")
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
#initializing random forest classifier
rf=RandomForestClassifier(n_estimators=1,random_state=42)
```

```python
#training the random forest model
rf.fit(x_train, y_train)
```

```python
y_pred1=rf.predict(x_test)
```

```python
accuracy2 = accuracy_score(y_test, y_pred1)
print(f"Accuracy by random forest classification = {(accuracy2 * 100):.2f}%")
```

```python
#decision tree classifier
dt=DecisionTreeClassifier(random_state=42)
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred2=dt.predict(x_test)
```

```python
accuracy3=accuracy_score(y_test, y_pred2)
print(f"Accuracy by decision tree classifier = {(accuracy3 * 100):.2f}%")
```
